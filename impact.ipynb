```python
def higherImpact(row):
    totalEmissions = row['Total_emissions']
    landChange = row['Land use change']
    
    if totalEmissions > 5 and landChange > 1:
        foodImpact = 'higher impact'
    else:
        foodImpact = 'lower impact'
    return foodImpact
    
impact['foodImpact'] = impact.apply(higherImpact, axis = 1)  
impact
        
```

```python
impact.groupby(['Food product', 'foodImpact']).mean().sort_values('foodImpact')
```
